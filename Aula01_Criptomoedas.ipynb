```python
import requests
import pandas as pd
```

```python
response = requests.get('https://api.wazirx.com/sapi/v1/tickers/24hr')

df = pd.read_json(response.text)

# df.describe() #quantidade, média, descvio padrão, minimo, mediana, primeiro quartio...
# df.info() #atributos, tipos de dados, quantos não nulos...


df ['variacao'] = df['openPrice'] - df['lastPrice'] #criando uma coluina de variação com a diferença abertura-final
variacao = df[['baseAsset', 'variacao']] #cria uma tabela somente com os campo baseAsset e variacao
variacao = variacao.sort_values('variacao', ascending = False ) #ordena a nova tabela pela variacao de forma descendente
variacao 
```
